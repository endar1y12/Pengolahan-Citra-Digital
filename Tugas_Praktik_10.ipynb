# Operator Robert

```python

import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
from skimage import io
from skimage.color import rgb2gray




roberts_v = np.array( [[ 0, 0, 0 ],
                      [ 0, 1, 0 ],
                      [ 0, 0,-1 ]] )

roberts_h = np.array( [[ 0, 0, 0 ],
                       [ 0, 0, 1 ],
                       [ 0,-1, 0 ]] )
url ="https://upload.wikimedia.org/wikipedia/en/a/a9/MarioNSMBUDeluxe.png"
img = io.imread(url)
img_gray =  rgb2gray(img)



vertical = ndimage.convolve( img_gray, roberts_v )
horizontal = ndimage.convolve( img_gray, roberts_h )

img_robert = np.sqrt( np.square(horizontal) + np.square(vertical))

plt.figure()
plt.imshow(img,cmap =plt.cm.gray)
plt.figure()
plt.imshow(img_gray,cmap =plt.cm.gray)
plt.figure()
plt.imshow(img_robert,cmap =plt.cm.gray)
```

# Operator Prewitt

```python
# import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import io

url ="https://upload.wikimedia.org/wikipedia/en/a/a9/MarioNSMBUDeluxe.png"
img = io.imread(url)


# img = data.camera()
img_gray =  rgb2gray(img)

#prewwitt
Px = np.array([[-1,0,1],
                    [-1,0,1],
                    [-1,0,1]])
Py = np.array([[1,1,1],
                    [0,0,0],
                    [-1,-1,-1]])

[r,c] =np.shape(img_gray)
prewitt = np.zeros(shape =(r,c))

for i in range(r -2):
  for j in range (c -2):
    px = np.sum(np.multiply(Px, img_gray[i:i + 3, j:j + 3]))
    py = np.sum(np.multiply(Py, img_gray[i:i + 3, j:j + 3]))
    prewitt[i + 1, j + 1] = np.sqrt(px ** 2 + py ** 2)

# img_prewittx = cv2.filter2D(img, -1, kernelx)
# img_prewitty = cv2.filter2D(img, -1, kernely)
# img_prewitt = img_prewittx + img_prewitty

plt.figure()
plt.imshow(img)
plt.figure()
plt.imshow(img_gray,cmap = plt.cm.gray)
plt.figure()
plt.imshow(prewitt,cmap = plt.cm.gray)
# plt.figure()
# plt.imshow(img_prewitty,cmap = plt.cm.gray)
# plt.figure()
# plt.imshow(img_prewitt,cmap = plt.cm.gray)
```

# Operator Sobel

```python
# import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import io
from skimage.color import rgb2gray

url ="https://upload.wikimedia.org/wikipedia/en/a/a9/MarioNSMBUDeluxe.png"
img = io.imread(url)

# img = data.camera()
img_gray = rgb2gray(img)
#sobel
# img_sobelx = cv2.Sobel(img,cv2.CV_8U,1,0,ksize=5)
# img_sobely = cv2.Sobel(img,cv2.CV_8U,0,1,ksize=5)
# img_sobel = img_sobelx + img_sobely
Gx = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
Gy = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
[r,c] =np.shape(img_gray)
sobel = np.zeros(shape =(r,c))

for i in range(r -2):
  for j in range (c -2):
    gx = np.sum(np.multiply(Gx, img_gray[i:i + 3, j:j + 3]))
    gy = np.sum(np.multiply(Gy, img_gray[i:i + 3, j:j + 3]))
    sobel[i + 1, j + 1] = np.sqrt(gx ** 2 + gy ** 2)

plt.figure()
plt.imshow(img)
plt.figure()
plt.imshow(img_gray,cmap =plt.cm.gray)
plt.figure()
plt.imshow(sobel,cmap =plt.cm.gray)
# plt.figure()
# plt.imshow(img_sobelx,cmap=plt.cm.gray)
# plt.figure()
# plt.imshow(img_sobely,cmap=plt.cm.gray)
# plt.figure()
# plt.imshow(img_sobel,cmap=plt.cm.gray)
```

# Operator Laplacian

```python
import cv2
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
# from skimage.filters import threshold_mean

url ="https://upload.wikimedia.org/wikipedia/en/a/a9/MarioNSMBUDeluxe.png"
img = io.imread(url)

# img = data.camera()
img_gray = rgb2gray(img)

# blur = cv2.medianBlur(img_gray,5)

laplacian = cv2.Laplacian(img_gray,cv2.CV_64F)


plt.figure()
plt.imshow(img)
plt.figure()
plt.imshow(img_gray,cmap=plt.cm.gray)
# plt.figure()
# plt.imshow(blur,cmap=plt.cm.gray)
plt.figure()
plt.imshow(laplacian,cmap=plt.cm.gray)
```
