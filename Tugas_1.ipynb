```python
from skimage import io
import matplotlib.pyplot as plt
from skimage.filters import threshold_mean
import numpy as np

img = io.imread("https://th.bing.com/th/id/R.894f3c57dea15181abbb1a285f196b00?rik=arRZYBl1wviOXA&riu=http%3a%2f%2fwww.e-architect.co.uk%2fimages%2fjpgs%2fshanghai%2fshanghai_tower_pirages281108_2.jpg&ehk=lSVF%2b2LIiCxaRdR4AFrvKyBlUcIWHrJzWjk0yaXviW4%3d&risl=&pid=ImgRaw&r=0")
plt.figure()
plt.imshow(img)


img_red = img[:,:,0]
img_green = img[:,:,1]
img_blue = img[:,:,2]


#melihat kanal warna Red
plt.figure()
plt.imshow(img_red,cmap=plt.cm.Reds)
#melihat histogram Red
plt.figure()
plt.hist(img_red.flatten(),bins=256)

#melihat kanal warna Blue
plt.figure()
plt.imshow(img_blue,cmap=plt.cm.Blues)
#melihat histogram Blue
plt.figure()
plt.hist(img_blue.flatten(),bins=256)

#melihat kanal warna Green
plt.figure()
plt.imshow(img_green,cmap=plt.cm.Greens)
#melihat histogram Green
plt.figure()
plt.hist(img_green.flatten(),bins=256)

#binerisasi

def binerisasi(img,threshold) :
  img_q = np.zeros((img.shape[0],img.shape[1]))
  for r in range(0,img.shape[0]):
    for c in range (0,img.shape[1]):
      if img[r,c] < threshold:
        img_q[r,c]= 0
      else:
        img_q[r,c] = 255

  return img_q

#Binerisasi Red
print("Binerisasi Red")
plt.figure( )
threshold = threshold_mean(img_red)
print(threshold)
img_libbiner =binerisasi(img_red,threshold)
plt.figure()
plt.imshow(img_libbiner,cmap=plt.cm.gray)

#Binerisasi Blue
print("Binerisasi Blue")
plt.figure( )
threshold = threshold_mean(img_blue)
print(threshold)
img_libbiner =binerisasi(img_blue,threshold)
plt.figure()
plt.imshow(img_libbiner,cmap=plt.cm.gray)

#Binerisasi Green
print("Binerisasi Green")
plt.figure( )
threshold = threshold_mean(img_green)
print(threshold)
img_libbiner =binerisasi(img_green,threshold)
plt.figure()
plt.imshow(img_libbiner,cmap=plt.cm.gray)


```
